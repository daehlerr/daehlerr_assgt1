The kNN classifier consists of two stages:

- During training, the classifier takes the training data and simply remembers it
- During testing, kNN classifies every test image by comparing to all training images and transfering the labels of the k most similar training examples
- The value of k is cross-validated

In this exercise you will implement these steps and understand the basic Image Classification pipeline, cross-validation, and gain proficiency in writing efficient, vectorized code.

```python
# Tinkering with broadcasting.
# Goal: To compute l2 (ell-two) norm of differences between pictures in
#  a test collection and pictures in a training collection.
#  ASSUME: Each image is a separate, single row of length
#  width*height
Test_x = np.array([[1.0, 2.0, 3.0, 4.0], 
                   [1.5, 2.5, 3.5, 4.5],
                   [1.7, 2.7, 3.7, 4.7]
                  ])
Train_x = np.array([[5.0,6.0, 7.0,8.0],
                    [0.5,0.6, 0.7,0.8]
                  ])
print '    Test data shape: ', Test_x.shape
print 'Training data shape: ', Train_x.shape
nbr_test_images = Test_x.shape[0]
nbr_train_images = Train_x.shape[0]
nbr_image_pairs = nbr_test_images*nbr_train_images

nbr_pixels_per_image = Test_x.shape[1]

print 'Nbrs of test images: ', nbr_test_images,
print ', training images: ', nbr_train_images,
print ', pixels per image: ', nbr_pixels_per_image

##  We must compare only matching pixels:
##      Test_Image[i]pixel[x,y]
##  with Training_Image[j]pixel[x,y]  (The same x,y in both.)
##  

print 'Test_x: '
print Test_x
print 'Train_x: '
print Train_x
image_diff = Test_x[0,:] - Train_x[0,:]
print 'Image[0,0] Difference: '
print image_diff

test_image = Test_x[0]
print 'test_image : '
# print '  element type: ', test_image[0].type
print '  test_image[] shape: ', test_image.shape
print '  dtype: ', test_image.dtype
print '  array: '
print test_image

## Try reshaping Test as (nbr_test_images, 1, nbr_pixes_per_image)
##  then broadcast against Train.  The result should be
##  (nbr_test_images, nbr_train_images, nbr_pixels_per_image),
##  which is promising.
Test_xx = Test_x.reshape(nbr_test_images,1,nbr_pixels_per_image)
print 'After reshaping, Test_xx:'
print '  shape: ', Test_xx.shape

diff = Test_xx - Train_x

print 'Broadcast difference:'
print '  shape: ', diff.shape
print diff

## Broadcast product of matrix with itself squares each element.
diff2 = diff*diff
print '(elements of diff) squared:'
print diff2

## Now take dot product of each image-diff-squared with
##  an image-length vector of all ones:
one_v = np.ones(nbr_pixels_per_image)
sums_of_squares = diff2.dot(one_v)
print 'Sums of squares: '
print '  shape: ', sums_of_squares.shape
print sums_of_squares

dists = np.sqrt(sums_of_squares)
print 'Distances: '
print '  shape: '
print dists

## Try raising an exception (to see how it works):
## raise AssertionError('This is a test of exception-handling.')
##
## print '*** We should not get here.'
```

We would now like to classify the test data with the kNN classifier. Recall that we can break down this process into two steps: 

1. First we must compute the distances between all test examples and all train examples. 
2. Given these distances, for each test example we find the k nearest examples and have them vote for the label

Lets begin with computing the distance matrix between all training and test examples. For example, if there are **Ntr** training examples and **Nte** test examples, this stage should result in a **Nte x Ntr** matrix where each element (i,j) is the distance between the i-th test and j-th train example.

```python
# Now lets speed up distance matrix computation by using partial vectorization
# with one loop. Implement the function compute_distances_one_loop and run the
# code below:
dists_one = classifier.compute_distances_one_loop(X_test)

# To ensure that our vectorized implementation is correct, we make sure that it
# agrees with the naive implementation. There are many ways to decide whether
# two matrices are similar; one of the simplest is the Frobenius norm. In case
# you haven't seen it before, the Frobenius norm of two matrices is the square
# root of the squared sum of differences of all elements; in other words, reshape
# the matrices into vectors and compute the Euclidean distance between them.
difference = np.linalg.norm(dists - dists_one, ord='fro')
print 'Difference was: %f' % (difference, )
if difference < 0.001:
  print 'Good! The distance matrices are the same'
else:
  print 'Uh-oh! The distance matrices are different'
```

```python
# Now implement the fully vectorized version inside compute_distances_no_loops
# and run the code
dists_two = classifier.compute_distances_no_loops(X_test)

# check that the distance matrix agrees with the one we computed before:
difference = np.linalg.norm(dists - dists_two, ord='fro')
print 'Difference was: %f' % (difference, )
if difference < 0.001:
  print 'Good! The distance matrices are the same'
else:
  print 'Uh-oh! The distance matrices are different'
```

```python
# Let's compare how fast the implementations are
def time_function(f, *args):
  """
  Call a function f with args and return the time (in seconds) that it took to execute.
  """
  import time
  tic = time.time()
  f(*args)
  toc = time.time()
  return toc - tic

two_loop_time = time_function(classifier.compute_distances_two_loops, X_test)
print 'Two loop version took %f seconds' % two_loop_time

one_loop_time = time_function(classifier.compute_distances_one_loop, X_test)
print 'One loop version took %f seconds' % one_loop_time

no_loop_time = time_function(classifier.compute_distances_no_loops, X_test)
print 'No loop version took %f seconds' % no_loop_time

# you should see significantly faster performance with the fully vectorized implementation
```

### Cross-validation

We have implemented the k-Nearest Neighbor classifier but we set the value k = 5 arbitrarily. We will now determine the best value of this hyperparameter with cross-validation.

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]
# k_choices = [1, 3]

X_train_folds = []
y_train_folds = []
################################################################################
# TODO:                                                                        #
# Split up the training data into folds. After splitting, X_train_folds and    #
# y_train_folds should each be lists of length num_folds, where                #
# y_train_folds[i] is the label vector for the points in X_train_folds[i].     #
# Hint: Look up the numpy array_split function.                                #
################################################################################
X_train_folds = np.array_split(X_train, num_folds)
y_train_folds = np.array_split(y_train, num_folds)

if (__debug__):
    print ' Length of X_train_folds: ', len(X_train_folds)
    print ' Length of y_train_folds: ', len(y_train_folds)
#endif debugging

# Fudge factor for comparing index values of different types:
epsilon = 0.01

# pass
################################################################################
#                                 END OF YOUR CODE                             #
################################################################################

# A dictionary holding the accuracies for different values of k that we find
# when running cross-validation. After running cross-validation,
# k_to_accuracies[k] should be a list of length num_folds giving the different
# accuracy values that we found when using that value of k.
k_to_accuracies = {}

################################################################################
# TODO:                                                                        #
# Perform k-fold cross validation to find the best value of k. For each        #
# possible value of k, run the k-nearest-neighbor algorithm num_folds times,   #
# where in each case you use all but one of the folds as training data and the #
# last fold as a validation set. Store the accuracies for all fold and all     #
# values of k in the k_to_accuracies dictionary.                               #
################################################################################
## pass

fold_range = arange(num_folds)

# Compute distances matrix for each fold.
image_size = X_train.shape[1]

for this_fold in fold_range:
    #  The test matrix is the current fold, 
    #  and the training matrix is everything else.
    this_test_matrix = X_train_folds[this_fold]
    this_test_labels = y_train_folds[this_fold]
    
    this_training_matrix = np.empty(shape=(0, image_size), dtype=X_train.dtype)
    this_training_labels = np.empty(shape=(0), dtype=y_train.dtype)
    for m in fold_range:
        if (m != this_fold):
            this_training_matrix = np.concatenate((this_training_matrix, X_train_folds[m]))
            this_training_labels = np.concatenate((this_training_labels, y_train_folds[m]))
        #endif not this fold
    #endfor m concatenating pieces

    # Now train the classifier with this set of training data.
    if (__debug__):
        print 'Fold #', this_fold
    #endif debugging
    classifier = KNearestNeighbor()
    classifier.train(this_training_matrix, this_training_labels)

    # Compute the distances matrix for this partition of 
    #  test and training data.
    #dists_two = classifier.compute_distances_two_loops(this_test_matrix)
    dists_zero = classifier.compute_distances_no_loops(this_test_matrix)

    #  Look at the performance of the classifier for different
    #  values of k.
    for this_k in k_choices:
        #this_y_test_pred = classifier.predict_labels(dists_two, this_k)
        if (__debug__):
            print
            print 'About to call predict_labels(dists_zero, this_k==', this_k, ')...'
        #endif debugging
        this_y_test_pred = classifier.predict_labels(dists_zero, this_k)

        # Compute and print the fraction of correctly predicted examples.
        #  Note the need to cast from integer to float before the division.
        #  ALSO: The values of y_test_pred and y_test may be of different
        #  types (int or float), so we need to allow for a fudge factor.
        #
        # num_correct = np.sum(y_test_pred == y_test)
        #
        if(__debug__):
            print 'Shape of this_y_test_pred: ', this_y_test_pred.shape
            print 'Shape of this_test_labels: ', this_test_labels.shape
        #endif debugging
        num_test = this_test_labels.shape[0]
        #num_correct = np.sum( (this_y_test_pred+np.multiply(-1, this_test_labels)) < epsilon)
        #num_correct = np.sum( (this_y_test_pred - this_test_labels) < epsilon)
        num_correct = np.sum( np.abs(this_y_test_pred - this_test_labels) < epsilon)
        accuracy = (1.0*num_correct) / num_test
        
        if (__debug__ and (this_k<6)):
            print 'fold==', this_fold, ' k==', this_k, ' accuracy==', accuracy
            
            print 'First few predictions: '
            temp_range = np.arange(5)
            for tr in temp_range:
                print 'this_y_test_pred[', tr, ']==', this_y_test_pred[tr], ' this_test_labels[', tr, ']==', this_test_labels[tr]
                print 'np.abs(this_y_test_pred[', tr, '] - this_test_labels[', tr, ']) ==', np.abs(this_y_test_pred[tr] - this_test_labels[tr])
                print 'np.abs(this_y_test_pred[', tr, '] - this_test_labels[', tr, ']) < epsilon : ', np.abs((this_y_test_pred[tr] - this_test_labels[tr])) < epsilon
            #endfor tr
            print 'num_correct==', num_correct
            print 'num_test==', num_test
            print 'accuracy==', accuracy
            print
            
        #endif debugging
        
        if ( not (this_k in k_to_accuracies) ):
            # Create a new dictionary entry with a single-element list.
            k_to_accuracies.update({this_k : [accuracy] })
        else:
            # Dictionary entry already exists, to append to it.
            k_to_accuracies[this_k].append(accuracy)
        #endif k is a new value -- else
        
    #endfor this_k
    
#endfor this_fold

if (__debug__):
    for tk in k_choices:
        if (tk in k_to_accuracies):
            print "k_to_accuracies[k==", tk, "]: ", k_to_accuracies[tk]
        else:
            print 'k==', tk, ' not found in k_to_accuracies.'
        #endif
    #endfor tk
#endif debugging

################################################################################
#                                 END OF YOUR CODE                             #
################################################################################
"""
# Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print 'k = %d, accuracy = %f' % (k, accuracy)
"""
```
